```python
# Pre-processing

from sklearn.preprocessing import StandardScaler

data_f = data_f.astype(float)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data_f)
```

```python
# Training with K-means 
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

results = []
k_inertia = []; 
k_sil = []; 
# K range
k_range = range(2,6)
# Training using K-means
for k in k_range:
    k_means_model = KMeans(n_clusters=k, random_state=42)
    k_labels = k_means_model.fit_predict(X_scaled)

    k_inertia.append(k_means_model.inertia_)
    k_sil.append(silhouette_score(X_scaled, k_labels))

# Printing all metrics
results_df = pd.DataFrame({
    "k": list(k_range),
    "Inertia": k_inertia,
    "Silhouette Score": k_sil
})
print(results_df)

best_k = list(k_range)[int(np.argmax(k_sil))]
print("Best k by silhouette:", best_k)
plt.figure(figsize=(6, 3))
plt.plot(list(k_range), k_inertia, marker="o")
plt.title("Elbow (Inertia)")
plt.xlabel("k")
plt.ylabel("Inertia")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.plot(list(k_range), k_sil, marker="o")
plt.title("Silhouette Score by k")
plt.xlabel("k")
plt.ylabel("Silhouette Score")
plt.tight_layout()
plt.show()
```

```python
# Centroids with distribution for each cluster plotted
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
centroids_pca = pca.transform(k_means_model.cluster_centers_)

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=k_labels, s=40, alpha=0.7)
plt.scatter(centroids_pca[:,0], centroids_pca[:,1],
            c='black', s=200, marker='X', label='Centroids')

plt.title(f"K-Means Clusters (k=5)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
# Showing the distribution of each cluster
plt.figure(figsize=(6,5))
for lab in np.unique(k_labels):
    mask = k_labels==lab
    plt.scatter(X_pca[mask,0], X_pca[mask,1], s=20, label=f'Cluster {lab}')
plt.title(f'KMeans k={best_k}'); plt.xlabel('PC1'); plt.ylabel('PC2'); plt.legend(); plt.tight_layout(); plt.show()
```

```python
# training with DBSCAN and trying many epsilon values to the how performance changes
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

eps_list = [0.5, 0.7, 0.9, 1.1, 1.3]
min_samples_list = [3, 4, 5]

results = []

for eps in eps_list:
    for ms in min_samples_list:
        dbscan = DBSCAN(eps=eps, min_samples=ms)
        dbscan_labels = dbscan.fit_predict(X_scaled)

        valid = dbscan_labels != -1
        n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)

        if valid.sum() > 0 and n_clusters > 1:
            dbscan_sil = silhouette_score(X_scaled[valid], dbscan_labels[valid])
        else:
            dbscan_sil_sil = -1  # invalid clustering

        results.append([eps, ms, n_clusters, dbscan_sil])

results_df = pd.DataFrame(results, 
                          columns=["eps", "min_samples", "Clusters Found", "Silhouette"])

print(results_df)
```

```python
# Training with Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering

aggloclus = AgglomerativeClustering(n_clusters=5)
agglo_labels = aggloclus.fit_predict(X_scaled)

agglo_sil = silhouette_score(X_scaled, agglo_labels)
print("Agglomerative Silhouette:", agglo_sil)
```

```python
# Comparing each model's performance
comparison = pd.DataFrame({
    "Algorithm": ["K-Means", "DBSCAN", "Agglomerative"],
    "Silhouette Score": [k_sil, dbscan_sil, agglo_sil]
})

print("\n--- ALGORITHM COMPARISON ---")
print(comparison)
```
